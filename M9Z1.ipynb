```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
film_base = pd.read_csv('tmdb_movies.csv')
film_base
```

```python
genre_base = pd.read_csv('tmdb_genres.csv')
genre_base
```

```python
#Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue) oraz średni budżet (budget) w danym roku dla filmów opublikowanych od 2010 
#(włącznie) do 2016 roku (włącznie). Następnie na tej podstawie stwórz wykres, w którym średnie przychody są wykresem kolumnowym,
#a średni budżet wykresem liniowym na tych samych osiach. Sformatuj odpowiednio oś X oraz oś Y. 
#Dodaj tytuł wykresu, oraz legendę, która znajduje się w prawym górnym rogu płótna, lecz poza obszarem osi.
```

```python
filtered_dates = film_base.loc[(film_base['release_date'] >= '2010-01-01')  # filtruję dane z dat od 2010 do 2016
                     & (film_base['release_date'] <= '2016-12-31')]
filtered_dates
```

```python
chart_data = filtered_dates[["budget", "revenue", "release_date"]]
chart_data
```

```python
chart_data['year'] = pd.DatetimeIndex(chart_data['release_date']).year
chart_data
```

```python
year_mean = chart_data.groupby('year').agg({'revenue': 'mean', 'budget': 'mean'}).reset_index()
year_mean
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(year_mean['year'], year_mean['revenue'], color='blue', label='Budget')
ax.plot(year_mean['year'], year_mean['budget'], color='red', label='Revenues',linewidth=3)
ax.set_title("Średni przychód i budżet filmu w latach 2010-2016")
ax.legend(loc='upper right', bbox_to_anchor=(1.20, 1))

def millions(x, pos):
    return f'{x * 1e-6:.1f} M'

ax.yaxis.set_major_formatter(plt.FuncFormatter(millions))

plt.xticks(year_mean['year'])
plt.show()
```

```python
#Baza filmów zawiera kolumnę z id gatunku (genre_id). Na tej podstawie połącz ze sobą bazę filmów z bazą gatunków, 
#tak aby w bazie filmów można było odczytać nazwę gatunku filmu.
```

```python
new_genre_base = (genre_base.rename({'Unnamed: 0': 'genre_id'}, axis='columns'))
new_genre_base
```

```python
combined_frames = (pd.merge(film_base, new_genre_base, on='genre_id'))
combined_frames
```

```python
#Jaki gatunek filmu z bazy pojawia się w niej najczęściej? Ile filmów tego gatunku znajduje się w bazie?
```

```python
count_genres = combined_frames['genres'].value_counts()
count_genres
```

```python
print(f"Najbardziej popularnym gatunkiem filmu jest Drama i pojawia się na liście {count_genres.max()} razy.")
```

```python
#Filmy, którego gatunku trwają średnio najdłużej (runtime)?
```

```python
print(combined_frames['genres'].head()) #sprawdzam czy czasem dany film nie ma po dwóch gatunków w kolumnie genres
runtime_genres = combined_frames.groupby('genres')['runtime'].mean().sort_values(ascending=False)
runtime_genres
```

```python
longest_one = runtime_genres.idxmax()
print(f'Gatunek o najdłuższym średnim czasie trwania to: {longest_one}')
```

```python
#Stwórz histogram czasu trwania filmów z gatunku, który cechuje się największym średnim czasem trwania.
```

```python
longest_genre_films = combined_frames[combined_frames['genres'] == longest_one]
longest_genre_films
```

```python
plt.figure(figsize=(10, 6))
plt.hist(longest_genre_films['runtime'].dropna(), bins=20, color='red', edgecolor ='blue')
plt.xlabel('Czas trwania (minuty)')
plt.ylabel('Liczba filmów')
plt.title(f'Histogram czasu trwania dla gatunku: {longest_one}')
plt.show()
```
